## Cross Correlations

There are a couple of different ways we can count spherelites but we are going to try to use a **cross correlation**.  In a cross correlation a **kernel** is slid across an image and multiplied.  This results in a transformed image which has peaks in areas similar to the origional image.

Let's see an example of this...

```python
#importing some functions and images to play with!
from skimage.data import astronaut
from scipy.signal import correlate
from skimage.color import rgb2gray
astronaut = rgb2gray(astronaut())
```

```python
# I am going to define the "Normalized cross correlation" function here.
from scipy.signal import correlate

def norm_cross_corr(image, kernel, **kwargs):
    """Computes the normalized cross correlation given an image and 
    a kernel.
    
    Parameters
    ----------
    image: np.array
        The image to compute the cross correlation of.
    kernel: np.array
        A kerel with the same dimensions as the image. This kernel is rastered 
        over the image and then multiplied at every position.
    
    Returns
    -------
    corr: np.array
        The normalized cross correlation at every point in the image.
    """
    norm_image = (image - np.mean(image)) / (np.std(image) * len(image))
    norm_kernel = (kernel - np.mean(kernel)) / (np.std(kernel))
    corr = correlate(norm_image, norm_kernel,**kwargs)
    return corr
```

```python
from skimage.feature import peak_local_max
```

```python
#lets just look for the astronaut's head
kernel = astronaut[0:175,125:300]
fig, axs = plt.subplots(1,3,figsize=(10,5))

#plotting the image
axs[0].imshow(astronaut, cmap="gray")
axs[0].set_title("Image")
# plotting the Kernel
axs[1].imshow(kernel, cmap="gray")
axs[1].set_title("Kernel")

#calculating the cross correlation
corr = norm_cross_corr(astronaut, kernel,mode="same")

axs[2].imshow(corr)
axs[2].set_title("Normalized Cross-Correlation")
peaks = peak_local_max(corr,threshold_rel=.95,min_distance=10)
axs[2].scatter(peaks[:,1],peaks[:,0])
axs[0].scatter(peaks[:,1],peaks[:,0])
plt.show()
```

### Figure 2:

Note in the images above that given the image of the astronauts head we can compute the cross correlation and easily find the face in the image just by finding the most intense point!

Of course if you try this on a different face it will not work very well but for finding one feature this works pretty well.

We also normalized the image.  This basically means that we care less about intensity in the image and care more about mataching variance or features in the image.

### Spherulite

Spheruilte in polymers are semicrystalline regions in the polymers which are alligned. When we apply a polarized light they create a "maltese cross" which creates a very interesting feature for proforming image processing on.

Try this out on your own and see how well you can find each of the crosses in the image.

Don't be afraid to try something new or different as well!

A good resource is https://scikit-image.org/docs/stable/auto_examples/index.html



```python
import skimage
import imageio
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage.feature import corner_fast, corner_peaks
```

```python
spherulite_ex = imageio.imread("Spheruliten.jpg")
```

```python
plt.figure(figsize=(15,15))
plt.imshow(spherulite_ex)
# convert the image to a gray scale image
greyscale_spherulite_ex = rgb2gray(spherulite_ex) 
```

```python
# TODO: Define a kernel around one of the crosses
kernel = greyscale_spherulite_ex[200:280,50:125] # taking just one cross
plt.figure()
plt.imshow(kernel)
```

```python
# TODO: Plot the cross correlation and try to find the center of (most) of the crosses.
#taking the normalized cross correlation
fig,axs = plt.subplots(1,2, figsize=(15,5))
axs[0].imshow(greyscale_spherulite_ex)

#calculating the cross correlation
cc = correlate(greyscale_spherulite_ex,kernel, mode="same")
axs[1].imshow(cc)

#finding the centers
peaks = skimage.feature.peak_local_max(cc,threshold_rel=.5,min_distance=10)
axs[0].scatter(peaks[:,1],peaks[:,0], c="w")
axs[1].scatter(peaks[:,1],peaks[:,0],c="w")
axs[0].set_title("Spherurlites with Centers Identified")
axs[1].set_title("Normalized Cross Correlation with Centers Identified")
plt.show()
```

```python
# give to students
from scipy.spatial import ConvexHull
def vornoi_area(voronoi):
    """ Calculates the area of a set of voronoi tellasaltions in nubmer of 
    pixels.  Gives nan for edges.
    """
    vol = np.zeros(voronoi.npoints)
    for i, reg_num in enumerate(voronoi.point_region):
        indices = voronoi.regions[reg_num]
        if -1 in indices: # some regions can be opened
            vol[i] = np.nan
        else:
            vol[i] = ConvexHull(voronoi.vertices[indices]).area
    return vol
```

```python
# TODO Make a histogram of the size of the spherulites in size vs the number
#of pixels for each voroni region.
area = vornoi_area(vor)
plt.hist(area, range=(0,1000),bins=30)
plt.ylabel("Number of Spherulites")
plt.xlabel("Size of the regions in Pixels")
```
